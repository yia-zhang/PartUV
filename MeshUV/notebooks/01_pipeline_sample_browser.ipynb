# Pipeline Sample Browser
按 UID / selection rank / 随机种子查看单个 accepted 样本。只读。

```python
import sys, os
sys.path.insert(0, os.path.abspath("../src"))
from meshuv.visualization import *
import matplotlib.pyplot as plt

DATASET_ROOT = "datasets/processed/MeshUV-TD-PseudoGT-MVP-v0"
UID = None        # 例 "00001dbc99db"
RANK = None       # 例 0
SEED = 0          # UID/RANK 均为 None 时随机
root = resolve_root(DATASET_ROOT)
index = load_index(root) if root else []
rec = pick_object(index, uid=UID, rank=RANK, seed=SEED)
item = open_sample(root, rec) if rec else None
print("对象:", rec and rec["object_id"])
```

## 样本全景(缺失字段会显示提示而非崩溃)

```python
if item:
    fig = plt.figure(figsize=(15, 9))
    show_basecolor(item, fig.add_subplot(2, 3, 1))
    show_mesh_preview(item, fig.add_subplot(2, 3, 2, projection="3d"))
    show_chart_segmentation(item, fig.add_subplot(2, 3, 3))
    show_demand_heatmap(item, fig.add_subplot(2, 3, 4))
    show_density_heatmap(item, fig.add_subplot(2, 3, 5))
    show_packed_atlas(item, fig.add_subplot(2, 3, 6))
    plt.tight_layout(); plt.show()
else:
    print("无可显示样本")
```
