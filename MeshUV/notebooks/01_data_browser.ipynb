# 01 Data Browser — Clean V1
只读; UID/rank/seed 三选一。

```python
%matplotlib inline
import sys, os
sys.path.insert(0, os.path.abspath("../src"))
import matplotlib.pyplot as plt
from meshuv.data.dataset import CleanDataset
from meshuv.visualization import pipeline_views as V

DATASET = "processed/clean_v1"      # 相对 MESHUV_DATA_ROOT(env 可覆盖)
root = V.resolve(DATASET)
ds = CleanDataset(root, expose_diagnostics=True) if root else None
print("objects:", ds and len(ds))
```

```python
UID = None; RANK = None; SEED = 0
i = V.pick(ds, uid=UID, rank=RANK, seed=SEED)
item = ds[i]
V.object_summary(item)
print("warnings:", __import__("json").load(open(f"{root}/objects/{item['object_id']}/status.json")).get("warnings", []))
```

```python
fig = plt.figure(figsize=(16, 8))
V.show_basecolor(item, fig.add_subplot(2, 3, 1))
V.show_mesh(item, fig.add_subplot(2, 3, 2, projection="3d"))
V.show_charts(item, fig.add_subplot(2, 3, 3))
V.show_coverage(item, fig.add_subplot(2, 3, 4))
V.show_signal(item, fig.add_subplot(2, 3, 5))
V.show_target(item, fig.add_subplot(2, 3, 6))
plt.tight_layout(); plt.show()
```
